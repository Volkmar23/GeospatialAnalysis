## 2. Calcular resumen

```python
import os
import pandas as pd
import numpy as np

path_results = r"C:\Users\Usuario\OneDrive - INTERCONEXION ELECTRICA S.A. E.S.P\01_CC_DOC\Notebooks_dowjones\Landslide\cache_results\Resultados_CIGRE.csv"

activos_dem = pd.read_csv(path_results)
activos_dem = activos_dem.reset_index(names = 'id')

labels = [ 'Baja', 'Intermedia', 'Alta', 'Muy Alta']
cols_melt = activos_dem.filter(like ='index')
# Melting the DataFrame
df_melted = activos_dem.melt(id_vars=[col for col in activos_dem.columns if col not in cols_melt], var_name='CODE', value_name='Index')
df_melted.loc[:, ['Escenario','horizonte']] =   df_melted.CODE.str.extract(r"(?P<Escenario>Historical|SSP2 4.5|SSP3 7.0|SSP5 8.5)_(?P<horizonte>\d+-\d+)_index")


for tower_id, df_id in df_melted.groupby('id'):
    # Get the index of the historical row and its value
    hist_idx = df_id[df_id['Escenario'] == 'Historical'].index[0]

    hist_value = df_id.loc[hist_idx, 'Index']

    # Get the indices of the future scenarios (non-historical)
    future_idx = df_id[df_id['Escenario'] != 'Historical'].index

    delta_change = df_melted.loc[future_idx, 'Index'] - hist_value
    df_melted.loc[future_idx, 'delta_change'] = delta_change
    
    # Compute relative change and assign it directly
    df_melted.loc[future_idx, 'relative_change'] = (
                                                    delta_change
                                                        ) / hist_value

df_melted_historical = df_melted.loc[df_melted.Escenario != 'Historical'].reset_index(drop = True)
# Sub-mask for valid slope (>=15) to classify by relative change
mask_high_sucptibility = (df_melted_historical.loc[:,'suceptilibty'] >= 3) & (df_melted_historical.loc[:,'Pendiente_12_5m_30m'] >= 20)

# First assign label 6 to the low slope but susceptible locations
df_melted_historical.loc[~ mask_high_sucptibility, 'landslide_label'] = 'Muy baja'
```

### Label risk assestment near future.

```python
near_future = df_melted_historical.loc[df_melted_historical.horizonte.isin( ['2020-2040', '2040-2060'])]

mask_high_sucptibility_near = near_future.landslide_label.isna()
values_gt_zero = near_future.relative_change  > 0
values_less_zero = ~values_gt_zero

mask_greter = values_gt_zero.values & mask_high_sucptibility_near.values
mask_lesser = values_less_zero.values & mask_high_sucptibility_near.values


bins = [- np.inf ] + list(np.quantile(near_future.loc[mask_greter].relative_change, q = [0.45,0.8,0.9])) + [np.inf]


near_future.loc[mask_greter, 'landslide_label'] = pd.cut(  near_future.loc[mask_greter].relative_change.values ,bins = bins,labels =labels) 
near_future.loc[mask_lesser, 'landslide_label'] = 'Baja'
```

```python
acciones_itco = {'Revegetalización':	 15_000,
                'Obra Menor (Trinchos, Cunetas, etc)':	 35_000,
                'Monitoreo':	 1_500,
                'Obras mayores (Recimentaciones y muros)':	 200_000,
                'Obras mayores (Variantes)':	 1_500_000,
                'Revegetalización':	 15_000,
                }
    
acciones_gaia = {'Techos Verdes':	 14_710,
                'Revision y Mantenimiento':	 8_185,
                'Refuerzo Estructural':	 14_235,
                'Muros de contención natural':	 23_825	,
                }

labels = [('Alta',0.1), ('Muy Alta',1) ]
thrshold = 200 
set_acciones  = [('SOLUCIONES ITCO' , acciones_itco) , ('SOLUCIONES GAIA', acciones_gaia)]

## PORCENTAJE DE OBRAS ESPECIALES
pct_obras_especiales = 0.05

total_a_interventir_usd = [ ]
global_agregados = [ ]
labe_risk_recorder = [ ] 

name_columns = ['Soluciones' ,'Acciones', 'Escenario',
                'Horizonte', 'Porcentajes USD','TORRES INTERVENIDAS' ]
name_columns_data_2 = ['Soluciones', 'Escenario','Horizonte' ,
                       'TOTAL A INVERTIR PREVENCION USD', 'TORRES A INTERVENIR' ,
                       'TORRES CON MAS DE UNA INTERVENCION' ]
 
for (escenario , horizonte) , df in coberturas_cruve.groupby(['Escenario' , 'horizonte']):
    
    amount_towers = { }
    torres_a_intervenir = 0

    for label,pct_intervencio in labels:
    
        cantidad = df.landslide_label.value_counts().to_dict().get(label,0)
        
        pct_intervencio = pct_intervencio if cantidad > thrshold else 1
        producto_ = cantidad * pct_intervencio
        torres_a_intervenir += producto_
        amount_towers.setdefault(label , producto_ ) 

        etiqueta_cols = (label ,cantidad,escenario,  horizonte)
        
        labe_risk_recorder.append( etiqueta_cols )
     

    for idx,(label_acciones , dict_acciones) in enumerate(set_acciones,start = 1):
        
        torres_intervenidas_totales = 0

        ## TOTAL A INVERTIR PREVENCION
        total_a_intervenir_prevencion_usd = 0
            
        for risk_label , amount_label in amount_towers.items():

            for accion_i,costo_usd in dict_acciones.items():
                
                if accion_i in [ 'Revegetalización', 'Techos Verdes']:

                    total_action_i = costo_usd * amount_label
                else:
                    total_action_i = pct_obras_especiales * costo_usd * amount_label
                
                total_a_intervenir_prevencion_usd += total_action_i
                torres_intervenidas = total_action_i / costo_usd
                torres_intervenidas_totales += torres_intervenidas

                tupla_results = (label_acciones ,accion_i, escenario , horizonte , total_action_i, torres_intervenidas)    
                total_a_interventir_usd.append(tupla_results)
        torres_con_mas_intervenciones = torres_intervenidas_totales - torres_a_intervenir
        tupla_more_towers  = (label_acciones, escenario , horizonte ,total_a_intervenir_prevencion_usd ,torres_a_intervenir ,torres_con_mas_intervenciones)
        global_agregados.append(tupla_more_towers)

table_financiera = pd.DataFrame(total_a_interventir_usd , columns=name_columns)
table_global = pd.DataFrame(global_agregados , columns=name_columns_data_2)
labe_risk_recorder = pd.DataFrame(labe_risk_recorder , columns = ['Riesgo', 'Cantidad' , 'Escenario','Horizonte' ] )

# table_financiera.to_csv(r"C:\Users\Usuario\OneDrive - INTERCONEXION ELECTRICA S.A. E.S.P\Varios\DowJones\Landslide\tablero\table_financiera.csv")
# table_global.to_csv(r"C:\Users\Usuario\OneDrive - INTERCONEXION ELECTRICA S.A. E.S.P\Varios\DowJones\Landslide\tablero\table_global.csv")
# labe_risk_recorder.to_csv(r"C:\Users\Usuario\OneDrive - INTERCONEXION ELECTRICA S.A. E.S.P\Varios\DowJones\Landslide\tablero\labe_risk_recorder.csv")

table_financiera.to_csv(r"C:\Users\Usuario\OneDrive - INTERCONEXION ELECTRICA S.A. E.S.P\01_CC_DOC\Notebooks_dowjones\Landslide\tablero\table_financiera.csv")
table_global.to_csv(r"C:\Users\Usuario\OneDrive - INTERCONEXION ELECTRICA S.A. E.S.P\01_CC_DOC\Notebooks_dowjones\Landslide\tablero\table_global.csv")
labe_risk_recorder.to_csv(r"C:\Users\Usuario\OneDrive - INTERCONEXION ELECTRICA S.A. E.S.P\01_CC_DOC\Notebooks_dowjones\Landslide\tablero\labe_risk_recorder.csv")
```
